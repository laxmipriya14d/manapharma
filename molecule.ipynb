```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('CSV.csv')

print("First few rows of the dataset:")
display(df.head())

print("\nData types and non-null counts:")
display(df.info())

print("\nSummary statistics for numerical columns:")
display(df.describe())

print("\nCheck for missing values:")
display(df.isna().sum())

df_clean = df.dropna().copy()


duplicates = df_clean.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

plt.figure(figsize=(6, 4))
plt.scatter(df_clean['MolecularWeight'], df_clean['XLogP'], color='teal', edgecolor='k')
plt.xlabel('Molecular Weight (g/mol)')
plt.ylabel('XLogP')
plt.title('Molecular Weight vs XLogP')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
plt.bar(df_clean['CID'].astype(str), df_clean['XLogP'], color='orange', edgecolor='k')
plt.xlabel('Compound CID')
plt.ylabel('XLogP')
plt.title('XLogP for Each Compound')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
plt.bar(df_clean['CID'].astype(str), df_clean['MolecularWeight'], color='purple', edgecolor='k')
plt.xlabel('Compound CID')
plt.ylabel('Molecular Weight (g/mol)')
plt.title('Molecular Weight for Each Compound')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```


**How trends in molecular properties support solvent selection in drug development:**

- *Molecular Weight* and *XLogP* (a measure of lipophilicity) are key factors in determining a compound's solubility.
- Compounds with high XLogP values are more soluble in nonpolar (organic) solvents, while those with low or negative XLogP values are more soluble in polar (aqueous) solvents.
- For example, ATP and L-Arginine have very low XLogP values (highly hydrophilic), so water or buffer solutions are preferred solvents.
- Aspirin and Paracetamol have moderate XLogP values, making them suitable for both aqueous and some non-aqueous formulations.
- Understanding these trends allows formulators to choose solvents that maximize drug solubility and stability, which is crucial for effective drug delivery.
